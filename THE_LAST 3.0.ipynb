```python
from numpy import empty, zeros, max
from pylab import show,plot,xlabel,ylabel,legend,grid
from scipy import array, arange, pi, sin, cos,exp
from timeit import timeit
tau=365 #dias del a√±o
A=10.
B=12.  
D=0.1  #difusividad Termica
L=20   #metros
N=100 #divisiones de la grilla 
a=L/N  #mallo espacial
h=0.1#paso en el timepo 


tend=3285
c=h*D/a**2
T=empty(N+1,float)
T[1:N]=10
T[N]=11
Tp=empty(N+1,float)
Tp[N]=11

t=0.0
for t in range(0,tend+1):
    T[0]=A+(B*sin((2*pi*t)/tau))
    Tp[0]=A+(B*sin((2*pi*t)/tau))
    for i in range(1,N):  
        Tp[i]=T[i]+c*(T[i+1]+T[i-1]-2*T[i])
    T,Tp=Tp,T 
    if t==tend-270:
        variable=str(t)
        leyenda='t='+variable
        plot(T,label=leyenda)
    if t==tend-180:
        variable=str(t)
        leyenda='t='+variable
        plot(T,label=leyenda)
    if t==tend-90:
        variable=str(t)
        leyenda='t='+variable
        plot(T,label=leyenda)
    if t==tend:
        variable=str(t)
        leyenda='t='+variable
        plot(T,label=leyenda)
legend()
grid("on")
xlabel("POSICION")
ylabel("TEMPERATURA")
show()
```

```python
L=0.01
D=4.25e-6
N=100
a=L/N
h=1e-4
epsilon=h/1000

Tlo=0.0
Tmid=20.0
Thi=50.0

t1=0.01
t2=0.1
t3=0.4
t4=1.0
t5=10.0
tend=t5+epsilon

T=empty(N+1,float)
T[0]=Thi
T[N]=Tlo
T[1:N]=Tmid
Tp=empty(N+1,float)
Tp[0]=Thi
Tp[N]=Tlo

t=0.0
c=h*D/a**2

while t<tend:
    
    for i in range(1,N):
        Tp[i]=T[i]+c*(T[i+1]+T[i-1]-2*T[i])
    T,Tp=Tp,T
    t+=h
    if abs(t-t1)<epsilon:
        plot(T)
    if abs(t-t2)<epsilon:
        plot(T)
    if abs(t-t3)<epsilon:
        plot(T)
    if abs(t-t4)<epsilon:
        plot(T)
    if abs(t-t5)<epsilon:
        plot(T)
xlabel("x")
ylabel("T")
show()
```
